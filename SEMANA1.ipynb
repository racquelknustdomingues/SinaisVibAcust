## Criação de Sinais Elementares

```python
# Definindo o intervalo de tempo
tempo = np.linspace(0, 10, 1000)  # Intervalo de 0 a 10 segundos com 1000 pontos

# Criando o sinal de rampa unitária
rampa_unitaria = tempo

# Plotando o sinal
plt.plot(tempo, rampa_unitaria)
plt.title('Sinal de Rampa Unitária')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Criando o sinal de degrau unitário
degrau_unitario = np.heaviside(tempo, 1)

# Plotando o sinal
plt.plot(tempo, degrau_unitario)
plt.title('Sinal de Degrau Unitário')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Definindo o intervalo de tempo
tempo = np.linspace(-1, 1, 1001)  # Intervalo de -1 a 1 segundo com 1000 pontos

# Criando a função do impulso unitário
def impulso_unitario(t):
    return np.where(t == 0, 1, 0)

# Calculando o impulso unitário para cada ponto no intervalo de tempo
impulso = impulso_unitario(tempo)

# Plotando o sinal
plt.plot(tempo, impulso)
plt.title('Sinal de Impulso Unitário')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Parâmetros do sinal
frequencia = 5  # Frequência do sinal em Hertz
amplitude = 1   # Amplitude do sinal
num_amostras = 1000  # Número de amostras

# Definindo o intervalo de tempo
tempo = np.linspace(0, 1, num_amostras)

# Gerando o sinal senoidal discreto
sinal_harmonico = amplitude * np.sin(2 * np.pi * frequencia * tempo)

# Plotando o sinal
plt.plot(tempo, sinal_harmonico)
plt.title('Sinal Harmônico Discreto')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Parâmetros do sinal
f_carrier = 1  # Frequência do sinal portador em Hz
f_modulation = 10  # Frequência do sinal modulante em Hz

# Definindo o intervalo de tempo
tempo = np.linspace(0, 1, 1000)

# Sinal modulante
carrier_signal = np.sin(2 * np.pi * f_modulation * tempo)

# Sinal portador
modulation_signal = np.sin(2 * np.pi * f_carrier * tempo)

# Sinal AM
am_signal = (modulation_signal) * carrier_signal

# Plotagem dos sinais
plt.plot(tempo, carrier_signal, label='Portador',alpha=0.5)
plt.plot(tempo, modulation_signal, label='Modulante',alpha=0.5)
plt.plot(tempo, am_signal, label='AM')
plt.title('Amplitude Modulada (AM)')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Parâmetros do sinal
Fs = 10000  # Frequência de amostragem (Hz)
duracao = 2  # Duração do sinal em segundos
f_modulante = 1  # Frequência do sinal modulante em Hz
f_c=10

# Tempo do sinal
t = np.arange(0, duracao, 1/Fs)

# Sinal modulante (senoide)
sinal_modulante = np.sin(2 * np.pi * f_modulante * t)

# Sinal FM
sinal_fm = np.cos(2 * np.pi * (sinal_modulante+f_c) * t)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t, sinal_modulante, label='Modulante',alpha=0.5)
plt.plot(t, sinal_fm, label='FM')
plt.title('Sinal FM com Modulação Senoidal de Frequência')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Parâmetros do sinal
Fs = 10000  # Frequência de amostragem (Hz)
duracao = 2  # Duração do sinal em segundos
f_modulante = 1  # Frequência do sinal modulante em Hz
f_c=10
f_carrier = 0.5  # Frequência do sinal portador em Hz

# Tempo do sinal
t = np.arange(0, duracao, 1/Fs)

modulation_signal = np.sin(2 * np.pi * f_carrier * tempo)

# Sinal modulante (senoide)
sinal_modulante = np.sin(2 * np.pi * f_modulante * t)

# Sinal FM
sinal_am_fm = (modulation_signal+1)*np.cos(2 * np.pi * (sinal_modulante+f_c) * t)



# Plotagem dos sinais
plt.plot(tempo, sinal_am_fm, label='AM+FM')
plt.title('AM+FM (Amplitude e Frequência Modulada)')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.legend()
# plt.xlim([0,0.2])
plt.grid(True)
plt.show()
```
